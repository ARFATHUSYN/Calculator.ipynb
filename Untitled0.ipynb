```python
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
column_names = ["sepal_length", "sepal_width", "petal_length", "petal_width", "class"]
iris_data = pd.read_csv(url, names=column_names)

print("First few rows of the Iris dataset:")
print(iris_data.head())

sepal_length_mean = iris_data["sepal_length"].mean()
sepal_length_median = iris_data["sepal_length"].median()
sepal_length_mode = iris_data["sepal_length"].mode().values[0]

print("\nMean sepal length:", sepal_length_mean)
print("Median sepal length:", sepal_length_median)
print("Mode sepal length:", sepal_length_mode)
```

```python
import numpy as np

array1 = np.array([[1, 2, 3],
                   [4, 5, 6]])

array2 = np.array([[7, 8, 9],
                   [10, 11, 12]])
vertical_concatenation = np.vstack((array1, array2))
horizontal_concatenation = np.hstack((array1, array2))
print("Array 1:")
print(array1)

print("\nArray 2:")
print(array2)

print("\nVertical Concatenation:")
print(vertical_concatenation)

print("\nHorizontal Concatenation:")
print(horizontal_concatenation)
```

```python
import pandas as pd

data = {
    'Date': ['2023-09-01', '2023-09-02', '2023-09-03', '2023-09-04', '2023-09-05'],
    'Product': ['A', 'B', 'A', 'C', 'B'],
    'Amount': [100, 150, 200, 75, 120]
}
df = pd.DataFrame(data)

df['Date'] = pd.to_datetime(df['Date'])
start_date = pd.to_datetime('2023-09-02')
end_date = pd.to_datetime('2023-09-04')

filtered_df = df[(df['Date'] >= start_date) & (df['Date'] <= end_date)]
print("Original DataFrame:")
print(df)

print("\nFiltered DataFrame (within the date range):")
print(filtered_df)
```

```python
import numpy as np

heights = np.array([63, 65, 68, 71, 72, 75, 60, 67, 70, 64])
weights = np.array([120, 140, 155, 185, 205, 215, 110, 150, 180, 130])
heights_variance = np.var(heights)
heights_std_dev = np.std(heights)
weights_variance = np.var(weights)
weights_std_dev = np.std(weights)

correlation_coefficient = np.corrcoef(heights, weights)[0, 1]

print("Descriptive Statistics:")
print("Height Variance:", heights_variance)
print("Height Standard Deviation:", heights_std_dev)
print("Weight Variance:", weights_variance)
print("Weight Standard Deviation:", weights_std_dev)
print("Correlation Coefficient between Height and Weight:", correlation_coefficient)
```

```python
import pandas as pd
data = {
    'Student': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Helen'],
    'Class': ['A', 'B', 'A', 'C', 'B', 'A', 'C', 'C'],
    'Score': [85, 92, 78, 89, 96, 88, 76, 90]
}
df = pd.DataFrame(data)
grouped = df.groupby('Class')
summary_statistics = grouped.agg({'Score': ['mean', 'median']})

summary_statistics.columns = ['Mean Score', 'Median Score']
print("Summary Statistics by Class:")
print(summary_statistics)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

iris = sns.load_dataset("iris")
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle("Exploratory Data Analysis of Iris Dataset")

sns.histplot(iris["sepal_length"], kde=True, ax=axes[0, 0], color='skyblue')
axes[0, 0].set_title("Distribution of Sepal Length")
axes[0, 0].set_xlabel("Sepal Length (cm)")
axes[0, 0].set_ylabel("Frequency")

sns.histplot(iris["sepal_width"], kde=True, ax=axes[0, 1], color='salmon')
axes[0, 1].set_title("Distribution of Sepal Width")
axes[0, 1].set_xlabel("Sepal Width (cm)")
axes[0, 1].set_ylabel("Frequency")

sns.histplot(iris["petal_length"], kde=True, ax=axes[1, 0], color='lightgreen')
axes[1, 0].set_title("Distribution of Petal Length")
axes[1, 0].set_xlabel("Petal Length (cm)")
axes[1, 0].set_ylabel("Frequency")

sns.histplot(iris["petal_width"], kde=True, ax=axes[1, 1], color='lightcoral')
axes[1, 1].set_title("Distribution of Petal Width")
axes[1, 1].set_xlabel("Petal Width (cm)")
axes[1, 1].set_ylabel("Frequency")
scatter = sns.scatterplot(data=iris, x="sepal_length", y="sepal_width", hue="species", ax=axes[1, 1])
scatter.set_title("Scatter Plot: Sepal Length vs. Sepal Width")

plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    'Month': pd.date_range(start='2022-01-01', periods=12, freq='M'),
    'Sales': np.random.randint(50000, 100000, size=12)
}
df = pd.DataFrame(data)
annual_total = df['Sales'].sum()
monthly_contributions = df['Sales'] / annual_total * 100

plt.figure(figsize=(8, 8))
plt.pie(monthly_contributions, labels=df['Month'].dt.strftime('%b'), autopct='%1.1f%%', startangle=140)
plt.title('Monthly Sales Contribution to Annual Total (2022)')
plt.show()
```
